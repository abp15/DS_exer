```python
## Key concepts
##f.read always reads from the point and pushes forward
##f.read(size) will read size characters and then pushes forward
##f.readline() will read will read 1 line at a time and then pushes forward
##f.readlines() will read will read all the lines at a time 
##f.tell() will tell will tell where the current position it is reading from 
##f.seek(number) will reset the current position to number 
```

```python
f = open('data/marks.csv','r')
print(f.name)
print(f.mode)
f.close()

## necessary to close the file
```

```python
with open('data/marks.csv','r') as f:
    size = 100
    f_contents = f.read(size)
    
    print(f_contents)
    f_contents = f.read(size)
    print(f_contents, end='')
    while len(f_contents) > 0:
        print(f_contents)
        f_contents = f.read(size)
        print(f.tell())
```

```python
# writing a file.
with open('data/test2.txt','w') as f:
    f.write('test')
    
```

```python
with open('data/test2.txt','r') as rf:
    with open('data/test2_copy.txt','w') as wf:
        for line in rf:
            wf.write(line)
   
```

```python
# for pictures you can't read in normal mode
#use binary mode to read and write 
with open('data/test2.txt','rb') as rf:
    with open('data/test2_copy.txt','wb') as wf:
        for line in rf:
            wf.write(line)
   
```

```python
#reading and writing in chunk sizes 
with open('data/test2.txt','rb') as rf:
    with open('data/test2_copy.txt','wb') as wf:
        chunk_size =4096
        rf_chunk = rf.read(chunk_size)
        while len(rf_chunk) > 0:
            wf.write(rf_chunk)
            rf_chunk = rf.read(chunk_size)
```
