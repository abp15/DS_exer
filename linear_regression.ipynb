```python
%matplotlib inline
import pandas as pd
import pandas_profiling
import seaborn as seabornInstance 

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
w_dat = pd.read_csv('data/Weather.csv')

w_dat.head()
```

```python
w_dat.shape
```

```python
w_dat.describe()
```

```python
pandas_profiling.ProfileReport(w_dat)
```

```python
plt.plot(w_dat['MinTemp'], w_dat['MaxTemp'],'g.')
plt.xlabel('Min Temp')
plt.ylabel('Max Temp')
plt.minorticks_on()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(w_dat['MaxTemp'])
```

```python
X = w_dat['MinTemp'].values.reshape(-1,1)
Y= w_dat['MaxTemp'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=0)
```

```python
regress = LinearRegression()
regress.fit(X_train, y_train) #training the algorithm
```

```python
print(regress.intercept_)
print(regress.coef_)
```

```python
Y_pred = regress.predict(X_test)
```

```python
new_df = pd.DataFrame({'Actual_MaxTemp':  y_test.flatten(),'Prdicted_MaxTemp': Y_pred.flatten()})
```

```python
new_df.head()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, Y_pred)))
print('R2 Score:', metrics.r2_score(y_test,Y_pred))
```

```python
from sklearn.compose import TransformedTargetRegressor

ttr = TransformedTargetRegressor(regress, func=np.log, inverse_func=np.exp)

ttr.fit(X_train, y_train)
ttr.predict(X_test)
```
